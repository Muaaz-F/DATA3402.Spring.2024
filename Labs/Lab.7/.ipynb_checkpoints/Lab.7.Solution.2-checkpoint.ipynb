```python

def compute_and_display_matrices(dataset, low_level_features, high_level_features):
    # Separating low-level and high-level features
    low_levels = dataset[low_level_features + ['signal']]
    high_levels = dataset[high_level_features + ['signal']]
    
    covariance_low = np.cov(low_levels.drop(columns='signal'), rowvar=False)
    correlation_low = np.corrcoef(low_levels.drop(columns='signal'), rowvar=False)
    
    covariance_high = np.cov(high_levels.drop(columns='signal'), rowvar=False)
    correlation_high = np.corrcoef(high_levels.drop(columns='signal'), rowvar=False)
    
    col_names_low = low_levels.columns.tolist()
    col_names_high = high_levels.columns.tolist()
    
    covariance_low_list = covariance_low.tolist()
    covariance_high_list = covariance_high.tolist()
    
    correlation_low_list = correlation_low.tolist()
    correlation_high_list = correlation_high.tolist()
    
    covariance_low_html = tab(covariance_low_list, tablefmt='html', headers=col_names_low)
    print("Covariance Matrix - Low-level Features:")
    display(HTML(covariance_low_html))
    
    correlation_low_html = tab(correlation_low_list, tablefmt='html', headers=col_names_low)
    print("\nCorrelation Matrix - Low-level Features:")
    display(HTML(correlation_low_html))
    
    covariance_high_html = tab(covariance_high_list, tablefmt='html', headers=col_names_high)
    print("\nCovariance Matrix - High-level Features:")
    display(HTML(covariance_high_html))
    
    correlation_high_html = tab(correlation_high_list, tablefmt='html', headers=col_names_high)
    print("\nCorrelation Matrix - High-level Features:")
    display(HTML(correlation_high_html))
```

```python
compute_and_display_matrices(df, low_level_features=['l_1_pT', 'l_1_eta', 'l_1_phi', 'l_2_pT', 'l_2_eta', 'l_2_phi', 'MET', 'MET_phi'],
                             high_level_features=['MET_rel', 'axial_MET', 'M_R', 'M_TR_2', 'R', 'MT2', 'S_R', 'M_Delta_R', 'dPhi_r_b', 'cos_theta_r1'])
```

### 5.3

```python
#  scenarios with different numbers of signal (N_S) and background (N_B) events

scenarios = {
    "Scenario 1": (10, 100),
    "Scenario 2": (100, 1000),
    "Scenario 3": (1000, 10000),
    "Scenario 4": (10000, 100000)
}
```

```python

#  a function to compute rates based on the selected scenario
def compute_rates(d, scenario, bins=100, sel=1):
    hist, bins_ = np.histogram(d, bins=bins, density=True)
    R = np.zeros_like(hist)
    if sel == 1:
        R = np.cumsum(hist[::-1])[::-1] * (bins_[1] - bins_[0])
    elif sel == 2:
        R = np.cumsum(hist) * (bins_[1] - bins_[0])
    elif sel == 3:
        hist, bins_ = np.histogram(np.abs(d - 1.1), bins=bins, density=True)
        R = np.cumsum(hist[::-1])[::-1] * (bins_[1] - bins_[0])
    elif sel == 4:
        hist, bins_ = np.histogram(np.abs(d - 1.1), bins=bins, density=True)
        R = np.cumsum(hist) * (bins_[1] - bins_[0])
    return R, bins_

```

```python

## function to compare significance for different scenarios


def compare_significance(df_sig, df_bkg, obs_name, scenarios, bins=100, log=False, ret_xc=False, sel=1):
    plt.figure(figsize=(10, 5))
    
    for name, (n_sig_expected, n_bkg_expected) in scenarios.items():
        TPR, bins_sig = compute_rates(df_sig[obs_name], (n_sig_expected, n_bkg_expected), bins=bins, sel=sel)
        FPR, bins_sig = compute_rates(df_bkg[obs_name], (n_sig_expected, n_bkg_expected), bins=bins_sig, sel=sel)
        
        n_sig_expected_prime = n_sig_expected * TPR
        n_bkg_expected_prime = n_bkg_expected * FPR
        
        sig = n_sig_expected_prime / np.sqrt(n_sig_expected_prime + n_bkg_expected_prime)
        
        if name == "1":
            linestyle = "-"
        elif name == "2":
            linestyle = "--"
        elif name == "3":
            linestyle = "-."
        else:
            linestyle = ":"
        
        plt.step(bins_sig[:-1], sig, label=name + " " + str((n_sig_expected, n_bkg_expected)), linestyle=linestyle)
    
    if log:
        plt.yscale("log")
    plt.xlabel('Threshold (x_c)')
    plt.ylabel('Significance')
    plt.title(f'Significance vs. Threshold for {obs_name}')
    plt.legend()
    plt.show()
```

```python
for var in VarNames[1:]:
    compare_significance(df_sig, df_bkg, var, scenarios)
```

```python
#  function to find the optimal selection criteria (x_c) for selected observables


def find_optimal_criteria(df_sig, df_bkg, obs_names, scenarios, bins=100, sel=1):
    optimal_criteria = {}
    for var in obs_names:
        x_c_values = []
        for name, (n_sig_expected, n_bkg_expected) in scenarios.items():
            TPR, bins_sig = compute_rates(df_sig[var], (n_sig_expected, n_bkg_expected), bins=bins, sel=sel)
            FPR, bins_sig = compute_rates(df_bkg[var], (n_sig_expected, n_bkg_expected), bins=bins_sig, sel=sel)
            n_sig_expected_prime = n_sig_expected * TPR
            n_bkg_expected_prime = n_bkg_expected * FPR
            sig = n_sig_expected_prime / np.sqrt(n_sig_expected_prime + n_bkg_expected_prime)
            max_i = np.argmax(sig)
            x_c_values.append(bins_sig[max_i])
        optimal_criteria[var] = x_c_values
    return optimal_criteria

```

```python
def compute_efficiency(signal_distribution, background_distribution, x_c):
    # c efficiency for signal and background based on the selection criteria x_c
    
    epsilon_S = np.sum(signal_distribution > x_c) / len(signal_distribution)
    epsilon_B = np.sum(background_distribution > x_c) / len(background_distribution)
    return epsilon_S, epsilon_B

def compute_expected_events(epsilon_S, epsilon_B, n_sig, n_bkg):
    # c expected number of signal and background events after selections
    N_S_prime = epsilon_S * n_sig
    N_B_prime = epsilon_B * n_bkg
    return N_S_prime, N_B_prime

def create_cut_flow_table(df_sig, df_bkg,
                          selected_observables, optimal_criteria, scenarios):
    # cut-flow table for each scenario and observable combination
    
    for scenario_name, (n_sig, n_bkg) in scenarios.items():
        print(f"Scenario: {scenario_name}")
        print("Observable\t\tε_S\t\tε_B\t\tN'_S\t\tN'_B\t\tσ_S'")
        print("-" * 70)
        for var in selected_observables:
            x_c = optimal_criteria[var][int(scenario_name) - 1]
            signal_distribution = df_sig[var]
            background_distribution = df_bkg[var]
            epsilon_S, epsilon_B = compute_efficiency(signal_distribution, background_distribution, x_c)
            N_S_prime, N_B_prime = compute_expected_events(epsilon_S, epsilon_B, n_sig, n_bkg)
            significance = N_S_prime / np.sqrt(N_S_prime + N_B_prime)
            print(f"{var}\t\t{epsilon_S:.4f}\t\t{epsilon_B:.4f}\t\t{N_S_prime:.2f}\t\t{N_B_prime:.2f}\t\t{significance:.4f}")
        print("\n")


```

```python
import random

def generate_uniform(N,x_min,x_max):
    out = []
    ### BEGIN SOLUTION
    out = [random.random() * (x_max - x_min) + x_min for _ in range(N)]
    # Fill in your solution here        
    
    ### END SOLUTION
    return out

def histogram(x,n_bins=10,x_min=None,x_max=None):
    ### BEGIN SOLUTION

    if x_min is None:
        x_min = min(x)
    if x_max is None:
        x_max = max(x)

    bin_size = (x_max - x_min) / n_bins

    hist = [0] * n_bins
    bin_edges = [x_min + i * bin_size for i in range(n_bins + 1)]

    for value in x:
        for i in range(n_bins):
            if x_min + i * bin_size <= value < x_min + (i + 1) * bin_size:
                hist[i] += 1
                break
        
    ### END SOLUTION

    return hist,bin_edges

def draw_histogram(x,n_bins,x_min=None,x_max=None,character="#",max_character_per_line=20):
    ### BEGIN SOLUTION

    hist,bin_edges = histogram(x,n_bins)

    for key, value in enumerate(hist):
        start_range = f"[{bin_edges[key]:.2f}, {bin_edges[key + 1]:.2f}]"
        print(f"{start_range.ljust(20)} : {'#' * value}")

    ### END SOLUTION
    return hist,bin_edges
```

```python
def compute_ROC(df_sig, df_bkg, selected_observables, successive_selections):
    # Initialize a dictionary to store ROC data for each observable
    roc_data = {}

    # Loop over each observable
    for var in selected_observables:
        # Extract the signal and background distributions for the current observable
        signal_distribution = df_sig[var]
        background_distribution = df_bkg[var]

        # Initialize lists to store FPR and TPR for each selection stage
        all_FPR = []
        all_TPR = []

        # Loop over each successive selection
        for selection_criteria in successive_selections:
            # Apply the selection criteria to the distributions
            selected_signal = signal_distribution
            selected_background = background_distribution
            for key, value in selection_criteria.items():
                selected_signal = selected_signal[selected_signal > value]
                selected_background = selected_background[selected_background > value]

            # Compute the ROC curve for the current selection stage
            FPR, TPR, _ = roc_curve([1]*len(selected_signal) + [0]*len(selected_background),
                                     list(selected_signal) + list(selected_background))
            
            # Compute the area under the ROC curve
            AUC = auc(FPR, TPR)

            # Store the FPR and TPR for the current selection stage
            all_FPR.append(FPR)
            all_TPR.append(TPR)

            # Store the ROC data for the current observable
            roc_data[var] = {"FPR": all_FPR, "TPR": all_TPR, "AUC": AUC}

    return roc_data
```

```python
h1 = generate_uniform(300, 10, 20)
h2 = generate_uniform(300, 10, 20)
h3 = generate_uniform(300, 10, 20)
uni = pd.DataFrame({"h1": h1, "h2": h2, "h3": h3})

successive_selections = [
    {"h1": 15, "h2": 12, "h3": 18},  #  1
    {"h1": 16, "h2": 13, "h3": 17},  #  2
    {"h1": 17, "h2": 14, "h3": 16}   # s 3
]

roc_data = compute_ROC(uni, uni, uni.columns, successive_selections)

#  ROC curves
plt.figure(figsize=(8, 6))
for var, data in roc_data.items():
    for i in range(len(data["FPR"])):
        plt.plot(data["FPR"][i], data["TPR"][i], label=f"{var} (Selection {i+1}), AUC: {data['AUC']:.2f}")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for Successive Selections')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def compute_rate(d,bins=100):
    hist,bins_=np.histogram(d,bins=bins,density=True)
    R = np.cumsum(hist[::-1])[::-1] * (bins_[1]-bins_[0])
    return R,bins_
```
