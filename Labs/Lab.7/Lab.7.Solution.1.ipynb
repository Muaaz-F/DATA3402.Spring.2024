#### Part D

```python

def compute_and_display_matrices(dataset, low_level_features, high_level_features):
    # Separating low-level and high-level features
    low_levels = dataset[low_level_features + ['signal']]
    high_levels = dataset[high_level_features + ['signal']]
    
    covariance_low = np.cov(low_levels.drop(columns='signal'), rowvar=False)
    correlation_low = np.corrcoef(low_levels.drop(columns='signal'), rowvar=False)
    
    covariance_high = np.cov(high_levels.drop(columns='signal'), rowvar=False)
    correlation_high = np.corrcoef(high_levels.drop(columns='signal'), rowvar=False)
    
    col_names_low = low_levels.columns.tolist()
    col_names_high = high_levels.columns.tolist()
    
    covariance_low_list = covariance_low.tolist()
    covariance_high_list = covariance_high.tolist()
    
    correlation_low_list = correlation_low.tolist()
    correlation_high_list = correlation_high.tolist()
    
    covariance_low_html = tab(covariance_low_list, tablefmt='html', headers=col_names_low)
    print("Covariance Matrix - Low-level Features:")
    display(HTML(covariance_low_html))
    
    correlation_low_html = tab(correlation_low_list, tablefmt='html', headers=col_names_low)
    print("\nCorrelation Matrix - Low-level Features:")
    display(HTML(correlation_low_html))
    
    covariance_high_html = tab(covariance_high_list, tablefmt='html', headers=col_names_high)
    print("\nCovariance Matrix - High-level Features:")
    display(HTML(covariance_high_html))
    
    correlation_high_html = tab(correlation_high_list, tablefmt='html', headers=col_names_high)
    print("\nCorrelation Matrix - High-level Features:")
    display(HTML(correlation_high_html))
```
